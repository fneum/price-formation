```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use(["bmh", "../../matplotlibrc"])
sns.set_palette("colorblind")
```

```python
colors = sns.color_palette("colorblind")
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# demand curve

c = "black"
x = [0, 30]
y = [100, 20]
ax.plot(x, y, color=c)

x = [30, 70]
y = [20, 20]
ax.plot(x, y, color=c)

x = [70, 77.5]
y = [20, 0]
ax.plot(x, y, color=c)

# a) demand

c = colors[0]
x = [0, 10]
y = [35, 35]
ax.plot(x, y, color=c)

x = [10, 10]
y = [35, 100]
ax.plot(x, y, color=c)

ax.text(12, 95, "price set by", color="#454545")
ax.text(12, 90, "demand", color=c)

# b) storage discharge

c = colors[1]
x = [0, 20]
y = [2, 2]
ax.plot(x, y, color=c)

x = [20, 20]
y = [2, 35]
ax.plot(x, y, color=c)

x = [20, 35]
y = [35, 35]
ax.plot(x, y, color=c)

x = [35, 35]
y = [35, 100]
ax.plot(x, y, color=c)

ax.text(37, 95, "price set by", color="#454545")
ax.text(37, 85, "storage\ndischarge", color=c)

# c) storage charge

c = colors[2]
x = [0, 50]
y = [1.25, 1.25]
ax.plot(x, y, color=c)

x = [50, 50]
y = [1.25, 35]
ax.plot(x, y, color=c)

x = [50, 65]
y = [35, 35]
ax.plot(x, y, color=c)

x = [65, 65]
y = [35, 100]
ax.plot(x, y, color=c)

ax.text(67, 95, "price set by", color="#454545")
ax.text(67, 85, "storage\ncharge", color=c)


# d) wind solar

c = colors[3]
x = [0, 85]
y = [0.5, 0.5]
ax.plot(x, y, color=c)

x = [85, 85]
y = [0.5, 35]
ax.plot(x, y, color=c)

x = [85, 100]
y = [35, 35]
ax.plot(x, y, color=c)

x = [100, 100]
y = [35, 100]
ax.plot(x, y, color=c)

ax.text(102, 95, "price set by", color="#454545")
ax.text(102, 85, "wind &\nsolar", color=c)


# labels

c = "#454545"

bbox = dict(facecolor="white", edgecolor=c, boxstyle="round,pad=0.2")

ax.text(21, 50, "1", color=c, bbox=bbox)
ax.text(76, 10, "1", color=c, bbox=bbox)
ax.text(40, 23, "2", color=c, bbox=bbox)
ax.text(10, 5, "3", color=c, bbox=bbox)

ax.text(95, 38, "4", color=c, bbox=bbox)
ax.text(60, 38, "4", color=c, bbox=bbox)
ax.text(30, 38, "4", color=c, bbox=bbox)
ax.text(5, 38, "4", color=c, bbox=bbox)

ax.text(5, 105, "2", color=c, bbox=bbox)
ax.text(10, 105, "storage charging bid", color=c)

ax.text(5, 113, "1", color=c, bbox=bbox)
ax.text(10, 113, "elastic demand", color=c)

ax.text(55, 105, "4", color=c, bbox=bbox)
ax.text(60, 105, "storage discharging bid", color=c)

ax.text(55, 113, "3", color=c, bbox=bbox)
ax.text(60, 113, "zero marginal cost wind & solar", color=c)


ax.set_ylim(0, 100)
ax.set_xlim(0, 120)

ax.set_ylabel("price [â‚¬/MWh]")
ax.set_xlabel("quantity [MW]")

ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()

for ext in ["png", "pdf"]:
    plt.savefig(f"../../figures/figure-price-setting.{ext}", bbox_inches="tight")
```
