```python
import matplotlib.pyplot as plt
import seaborn as sns
import pypsa
import pandas as pd
import numpy as np

plt.style.use(["bmh", "../../matplotlibrc"])
sns.set_palette("colorblind")

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100

from plot import get_price_duration
```

```python
voll_files = {
    1: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+1-voll+true.nc",
    4: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+4-voll+true.nc",
    16: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+16-voll+true.nc",
    32: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+32-voll+true.nc",
    70: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+70-voll+true.nc",
}
ns_voll = {k: pypsa.Network(v) for k, v in voll_files.items()}
```

```python
pwl_files = {
    1: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+1-elastic_pwl+default.nc",
    4: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+4-elastic_pwl+default.nc",
    16: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+16-elastic_pwl+default.nc",
    32: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+32-elastic_pwl+default.nc",
    70: "../../results/20240426-num-years/network_solved_LT-country+DE-number_years+70-elastic_pwl+default.nc",
}
ns_pwl = {k: pypsa.Network(v) for k, v in pwl_files.items()}
```

```python
pdc_voll = {k: get_price_duration(v) for k, v in ns_voll.items()}
pdc_pwl = {k: get_price_duration(v) for k, v in ns_pwl.items()}
```

```python
pdc_h2_voll = {k: get_price_duration(v, bus='hydrogen') for k, v in ns_voll.items()}
pdc_h2_pwl = {k: get_price_duration(v, bus='hydrogen') for k, v in ns_pwl.items()}
```

```python
voll_st_files = {
    "LT": "../../results/20240424-70a/network_solved_LT-country+DE-voll+true.nc",
    "ST": "../../results/20240424-70a/network_myopic_LT-country+DE-voll+true_ST-perfect_foresight+true.nc",
    "ST - capacities +1%": "../../results/20240424-70a/network_myopic_LT-country+DE-voll+true_ST-horizon+96-overlap+48-perturbation+1.01.nc",
    "ST - capacities -1%": "../../results/20240424-70a/network_myopic_LT-country+DE-voll+true_ST-horizon+96-overlap+48-perturbation+0.99.nc",
}
ns_voll_st = {k: pypsa.Network(v) for k, v in voll_st_files.items()}
```

```python
pwl_st_files = {
    "LT": "../../results/20240424-70a/network_solved_LT-country+DE-elastic_pwl+default.nc",
    "ST": "../../results/20240424-70a/network_myopic_LT-country+DE-elastic_pwl+default_ST-perfect_foresight+true.nc",
    "ST - capacities +1%": "../../results/20240424-70a/network_myopic_LT-country+DE-elastic_pwl+default_ST-horizon+96-overlap+48-perturbation+1.01.nc",
    "ST - capacities -1%": "../../results/20240424-70a/network_myopic_LT-country+DE-elastic_pwl+default_ST-horizon+96-overlap+48-perturbation+0.99.nc",
}
ns_pwl_st = {k: pypsa.Network(v) for k, v in pwl_st_files.items()}
```

```python
pdc_voll_st = {k: get_price_duration(v) for k, v in ns_voll_st.items()}
pdc_pwl_st = {k: get_price_duration(v) for k, v in ns_pwl_st.items()}
```

```python
fn = "../..//data/time_series_60min_singleindex_filtered.csv"
history = (
    pd.read_csv(fn, index_col=0)["DE_LU_price_day_ahead"]
    .dropna()
    .sort_values(ascending=False)
    .reset_index(drop=True)
)
history.index = np.arange(0, 100, 100 / len(history.index))
```

```python
pu_colors = sns.color_palette("rocket", 5)
cat_colors = sns.color_palette("colorblind")
lw = 1.25

fig, axs = plt.subplots(3, 2, figsize=(10, 10))

for i, (ys, pdc) in enumerate(pdc_voll.items()):
    label = f"{ys} year" if ys == 1 else f"{ys} years"
    pdc.plot(ax=axs[0,0], label=label, color=pu_colors[i], clip_on=False, linewidth=lw)

for i, (ys, pdc) in enumerate(pdc_pwl.items()):
    label = f"{ys} year" if ys == 1 else f"{ys} years"
    pdc.plot(ax=axs[0,1], label=label, color=pu_colors[i], clip_on=False, linewidth=lw)

for i, (ys, pdc) in enumerate(pdc_h2_voll.items()):
    label = f"{ys} year" if ys == 1 else f"{ys} years"
    pdc.plot(ax=axs[1,0], label=label, color=pu_colors[i], clip_on=False, linewidth=lw)

for i, (ys, pdc) in enumerate(pdc_h2_pwl.items()):
    label = f"{ys} year" if ys == 1 else f"{ys} years"
    pdc.plot(ax=axs[1,1], label=label, color=pu_colors[i], clip_on=False, linewidth=lw)

for i, (key, pdc) in enumerate(pdc_voll_st.items()):
    ls = ":" if key != "LT" else "-"
    pdc.plot(ax=axs[2,0], label=key, color=cat_colors[i], linestyle=ls, linewidth=lw)

for i, (key, pdc) in enumerate(pdc_pwl_st.items()):
    ls = ":" if key != "LT" else "-"
    pdc.plot(ax=axs[2,1], label=key, color=cat_colors[i], linestyle=ls, linewidth=lw)

for i in range(2):
    history.plot(ax=axs[0,i], label='historic\n(2015-2020)', color='#aaa', linewidth=lw, zorder=0.5)

axs[0,0].set_title("a) VOLL: electricity price duration curve", loc='left', fontsize='medium')
axs[0,1].set_title("b) PWL-elastic: electricity price duration curve", loc='left', fontsize='medium')

axs[1,0].set_title("c) VOLL: hydrogen price duration curve", loc='left', fontsize='medium')
axs[1,1].set_title("d) PWL-elastic: hydrogen price duration curve", loc='left', fontsize='medium')

axs[2,0].set_title("e) VOLL: price duration curve with perturbations", loc='left', fontsize='medium')
axs[2,1].set_title("f) PWL-elastic: price duration curve with perturbations", loc='left', fontsize='medium')

for i in range(2):
    axs[0,i].set_ylim(-100, 2200)
    axs[0,i].set_ylabel("Electricity Price [/MWh]")
    axs[0,i].set_yticks(range(0, 2001, 250), minor=True)

for i in range(2):
    axs[2,i].set_ylim(-100, 2200)
    axs[2,i].set_ylabel("Electricity Price [/MWh]")
    axs[2,i].set_yticks(range(0, 2001, 250), minor=True)

for i in range(2):
    axs[1,i].set_ylim(-50, 1100)
    axs[1,i].set_ylabel("Hydrogen Price [/MWh]")
    axs[1,i].set_yticks(range(0, 1000, 100), minor=True)

for i in range(3):
    for j in range(2):
        axs[i,j].set_xlabel("Fraction of time [%]")
        axs[i,j].axhline(0, linewidth=1, color='#bbb', zorder=0)
        axs[i,j].set_xlim(0, 100)
        axs[i,j].legend()

for ext in ['png', 'pdf']:
    plt.savefig(f"../../resources/figure-panel.{ext}", bbox_inches='tight', dpi=300)
```
